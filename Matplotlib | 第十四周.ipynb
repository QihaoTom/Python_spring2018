```python
plt.figure(figsize=(10,3),dpi=80)
t=np.arange(0.,4.,0.1)
plt.plot(t,t,color='red',linestyle='-',linewidth=3,label='line 1')
plt.plot(t,t+2,color='green',linestyle='',marker='*',label='line 2')
plt.plot(t,t**2,color='blue',linestyle='',marker='+',label='line 3')
plt.legend(loc='upper left')
plt.show()
```

```python
# plt.subplot(211) 行，列，索引
fig,(ax0,ax1)=plt.subplots(2,1)
ax0.plot(range(7),[3,4,7,6,2,8,9],color='r',marker='o')
ax0.set_title('subplot1')
plt.subplots_adjust(hspace=0.5)
ax1.plot(range(7),[5,1,8,2,6,9,4],color='green',marker='o')
ax1.set_title('subplot2')
plt.show()
```

```python
import tushare as ts
fig=plt.figure(1)
plt.figure(figsize=(30,10),dpi=100)
a=ts.get_hist_data('600100').head(50) 
b=a.iloc[:,:5] # loc为定位置，取前五列数据
(x,y)=(b.index,b['open'])
(x1,y1)=b.index,b['close'] # 为坐标对，['open']为取open这个变量
plt.plot(x,y,color='r',marker='o')
plt.plot(x1,y1,color='b',marker='o')
plt.legend(loc='upper right')
plt.title('stock')
plt.xlabel('date')
plt.ylabel('stock price')
plt.show()
fig.savefig('tushare.jpg')
```
